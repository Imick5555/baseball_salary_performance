```python
import pandas as pd
import requests
from bs4 import BeautifulSoup, Comment
import time

# ---------------- CONFIG ----------------
INPUT_CSV  = r"C:\Users\Isaac\OneDrive\Documents\fall 2025 semester\STAT 386\MLB_2018_2025_Cleaned.csv"
OUTPUT_CSV = r"C:\Users\Isaac\OneDrive\Documents\fall 2025 semester\STAT 386\MLB_2018_2025_with_Salaries.csv"
pd.set_option('display.max_colwidth', None)
# ---------------- HELPER FUNCTIONS ----------------
def find_table_in_comments(soup):
    """
    Recursively searches all comments in soup for <table id="br-salaries">
    Returns the table if found, else None
    """
    for comment in soup.find_all(string=lambda text: isinstance(text, Comment)):
        inner_soup = BeautifulSoup(comment, "html.parser")
        table = inner_soup.find("table", id="br-salaries")
        if table:
            return table
        # Recursive search in nested comments
        nested_table = find_table_in_comments(inner_soup)
        if nested_table:
            return nested_table
    return None

def get_salary_for_year(url, year):
    """
    Fetches the player's salary for a given year.
    Returns int salary if found, else None.
    """
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/124.0.0.0 Safari/537.36",
        "Referer": "https://www.google.com/",
    }
    try:
        resp = requests.get(url, headers=headers, timeout=15)
        if resp.status_code != 200:
            return None
    except requests.RequestException:
        return None

    soup = BeautifulSoup(resp.text, "html.parser")
    table = find_table_in_comments(soup)
    if not table:
        return None

    # Search rows for the matching year
    for row in table.select("tbody tr"):
        year_tag = row.find("th", {"data-stat": "year_ID"})
        salary_tag = row.find("td", {"data-stat": "Salary"})
        if year_tag and salary_tag and year_tag.get_text(strip=True) == str(year):
            salary_text = salary_tag.get_text(strip=True)
            salary_text = salary_text.replace("$", "").replace(",", "")
            try:
                return int(float(salary_text))
            except ValueError:
                return None
    return None

# ---------------- MAIN SCRIPT ----------------
df = pd.read_csv(INPUT_CSV)

df["Salary"] = 0                # integer version
df["Salary_formatted"] = ""     # pretty version

cache = {}
total = len(df)
print(f"Starting scraping {total} rows...")
base_url = "https://www.baseball-reference.com"
for idx, row in df.iterrows():
    if idx % 10 == 0:
        print(f"Row {idx}/{total} → Salaries found: {(df['Salary'] > 0).sum()}")

    link = row["Player_Link"]
    year = int(row["Year"])

    if pd.isna(link) or year not in range(2018, 2026):
        continue

    if link.startswith("/"):
        link = base_url + link
    key = (link, year)
    if key in cache:
        df.at[idx, "Salary"] = cache[key]
        df.at[idx, "Salary_formatted"] = f"${cache[key]:,}"
        continue

    salary = get_salary_for_year(link, year)
    if salary:
        df.at[idx, "Salary"] = salary
        df.at[idx, "Salary_formatted"] = f"${salary:,}"
        cache[key] = salary

    # polite scraping
    time.sleep(1)

# Save output
df.to_csv(OUTPUT_CSV, index=False)
print(f"Done! Salaries saved to {OUTPUT_CSV}")
```
